```python
import requests
from bs4 import BeautifulSoup
```

```python
url ="https://www.imdb.com/title/tt8946378/reviews/?ref_=tt_ururv_sm&sort=submission_date%2Cdesc"
```

```python
s = requests.session()
headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36'}
response = s.get(url, allow_redirects=True, timeout=10, headers=headers)
response.status_code
```

```python
soup = BeautifulSoup(response.text,"html.parser")
soup
```

```python
reviews = soup.find_all(attrs={'data-testid': 'review-card-parent'})
len(reviews)
```
